## Step 5: Brainstorm your Approach.
##### In bullet-list form, provide a proposed approach for evaluating your problem statement. This can be somewhat high-level, but start to think about ways you can massage the data for maximum efficacy. 

- Extract the data for each group from the melted table.  Essentially, this will result in two separate lists of the tracks.  Each list will contain track names multiple times for each genre.  Each time the track name appears on the list will represent a week that the track appeared on the chart.

```python
# Extracting data
rock = melted_df.ix[melted_df['genre'] == 'Rock']['track']
rap = melted_df.ix[melted_df['genre'] == 'Rap']['track']
print rock
print rap
```

```python
# Getting value counts for tracks
rck_trcks = pd.value_counts(rock)
rap_trcks = pd.value_counts(rap)
```

```python
# Getting lists of only the counts
rck_wks = [i for i in rck_trcks]
rap_wks = [i for i in rap_trcks]
print rck_wks
print rap_wks
```

```python
# Look at normality of each dataset since normality is an assumption of the t test.
plt.hist(rck_wks)
plt.show()
plt.hist(rap_wks)
plt.show()
# Neither appear to follow a normal distribution.
# Running a hypothesis test for normality, where the null hypothesis is that the data follow a normal distribution.
print stats.mstats.normaltest([i for i in rck_wks])
print stats.mstats.normaltest([i for i in rap_wks])
# Both tests give a p-value below 0.05, so normal distribution is followed in neither.
# Transform the data to fit a normal distribution
t_rck_wks = [i**.5 for i in rck_wks]
t_rap_wks = [i**.5 for i in rap_wks]
# Test transformed data
plt.hist(t_rck_wks)
plt.show()
plt.hist(t_rap_wks)
plt.show()
# The data certainly look closer to following a normal distribution.
print stats.mstats.normaltest(t_rck_wks)
print stats.mstats.normaltest(t_rap_wks)
# Both tests give a p-value above 0.05, showing that the data reasonably follow a normal distribution.
```

```python
# Since equal variances are also an assumption for the t test, I will check for this using the Levene test.
print stats.levene(t_rck_wks, t_rap_wks)
# The Levene test for equal variances gave a p-value above 0.05, so the null hypothesis that the variances are 
# equal will not be rejected.
# Testing the null hypothesis that genre has no effect on length of time spent on chart.
stats.ttest_ind(t_rck_wks, t_rap_wks, equal_var=True)
```

The t-test gave a p-value of 0.00496, which is below the significance level of 0.05.  This means that the null hypothesis will be rejected.  Genre does appear to have a significant effect on the length of time a track spends on the Billboard chart.
